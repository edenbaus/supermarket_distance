```python
def get_soup(url):
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
    return BeautifulSoup(requests.get(url,headers=headers).content,'html5lib')
```

```python
url = 'http://supermarketpage.com/state/{state}/'.format(state=state)
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
soup = BeautifulSoup(requests.get(url,headers=headers).content,'html5lib')

supermarket_url_dict = {supermarket.text.replace('&amp;','&').lower():supermarket['href'] for supermarket in soup.findAll('table')[0].findAll('a')}
```

```python
def get_store_listing_df(Store):
    # returns dataframe of list of locations and addresses for a given supermarket store from supermarketpage.com
    store = [x for x in  supermarket_url_dict.keys() if x.startswith(Store.lower()) == 1 or x.find(Store.lower()) != -1][0]
    table = """{}""".format(get_soup(supermarket_url_dict[store.lower()]).find('table'))
    df = pd.read_html(table)[0].iloc[1:].reset_index(drop=True).rename(columns={0:'Store #', 1:"Address", 2:"State", 3:"Zip", 4:"Phone"})
    df['Address'] = df['Address']#.map(lambda address: address[:-1].strip()) #chop off extra ending ','
    df['store'] =  store.title()
    return df
```
